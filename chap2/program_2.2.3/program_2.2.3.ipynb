```python
#导入使用的库
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

#输入序列的时域波形
delta_n = np.arange(21)
x_n = np.power(0.5,delta_n)

#系统在0~20区间的样值
b = np.array([1,1]) #分子
a = np.array([1,-0.25,0.5]) #分母
zi = signal.lfilter_zi(b,a)*0 #系统初始零状态
z1,_ = signal.lfilter(b,a,x_n,zi=zi) #系统输出

#系统零状态响应
delta = signal.unit_impulse(20)
z2,_ = signal.lfilter(b,a,delta,zi=zi) #零状态响应

#绘制信号及系统时域波形
fig,ax = plt.subplots(3,1,constrained_layout=True)
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
ax[0].stem(x_n,basefmt="") #输入序列的时域波形
ax[0].set_title("输入序列的时域波形");ax[0].set_xlabel('n')
ax[1].stem(z1,basefmt="") #系统0~20区间的样值
ax[1].set_title("系统0~20区间的样值");ax[0].set_xlabel('n')
ax[2].stem(z2,basefmt="") #系统零状态响应波形
ax[2].set_title("系统零状态响应波形");ax[0].set_xlabel('n')
fig.savefig('./diff_system_solution.png',dpi=500)
```
